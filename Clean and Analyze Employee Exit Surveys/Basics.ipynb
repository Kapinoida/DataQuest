- Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

## Identify Missing Values and Drop Unnecessary Columns

Since there is a value type that would alter our anaylsis, we can read the csv again and assign the NaN values to the '`Not Stated`' values.

```python
dete_survey = pd.read_csv('dete_survey.csv', na_values='Not Stated')

dete_survey.head()
```

You can see the values have changed to NaN in the Start Date columns in out quick view.

Next, we need to get rid of columns we don't need altogether. Let's list out the columns for each survey.

```python
print(dete_survey.columns)
print(tafe_survey.columns)
```

Let's get rid of many of the questions and topics for both surveys.

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49], axis=1)
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66], axis=1)

print(dete_survey_updated.columns)
print(tafe_survey_updated.columns)
```

There. Much easier to work with and gets rid of many things we don't need.

## Clean Column Names

When we combine these dataframes, we want to do al little work trying to clean it up afterwards. We can start by standardizing the column names that overlap between datasets.

Let's start with the `dete_survey_updated` dataframe.

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(' ', '_')

print(dete_survey_updated.columns)
```

Now, let's handle some of the names in the `tafe_survey_updated` dataframe. We will use a different tactic, as the column names are much different than what we want.

```python
mapping = {'Record ID': 'id',
           'CESSATION YEAR': 'cease_date',
           'Reason for ceasing employment': 'separationtype',
           'Gender. What is your Gender?': 'gender',
           'CurrentAge. Current Age': 'age',
           'Employment Type. Employment Type': 'employment_status',
           'Classification. Classification': 'position',
           'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
           'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'
          }
tafe_survey_updated = tafe_survey_updated.rename(mapping, axis=1)

print(tafe_survey_updated.columns)
```

Now, we are starting to see some overlapping column names.

## Filter the Data

Let's recall some of the questions we want to answer:

- Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?
- Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

We are primarily dealing with employees who have resigned, so we want to look for only those respondents that have resigned in the `separationtype` columns of each dataframe.

```python
dete_survey_updated['separationtype'].value_counts()
```

```python
tafe_survey_updated['separationtype'].value_counts()
```

`tafe_survey_updated` has a value of '`Resignation`' only, while `dete_survey_updated` had three options. Let's just change each option to just `Resignation`.

```python
dete_survey_updated['separationtype'] = dete_survey_updated['separationtype'].str.split('-').str[0]

dete_survey_updated['separationtype'].value_counts()
```
